<font color = "red">EXERCISE</font> <br><br>
Assign a numeric value to a variable DNA_seq_length.<br>
Write a composite condition to check if the variable is in the range 10 to 100.<br>
Write a composite condition to check if the variable is lower than 50 or higher than 1000.<br>
Write a composite condition to check if the variable is lower than 200 or higher than 1000 and if it can be divided by 3 or 9. <br>

````python
### Write your solution in the following cells
# generated with perplexity ai
# https://www.perplexity.ai/search/write-python-code-with-comment-VBuI50n.STO1s3Gk15j.4Q

# Assign a numeric value to the variable DNA_seq_length
DNA_seq_length = 75

# Composite condition to check if the variable is in the range 10 to 100 (inclusive)
in_range_10_to_100 = 10 <= DNA_seq_length <= 100

# Composite condition to check if the variable is lower than 50 or higher than 1000
lower_than_50_or_higher_than_1000 = DNA_seq_length < 50 or DNA_seq_length > 1000

# Composite condition to check if the variable is lower than 200 or higher than 1000
# and if it can be divided by 3 or 9
complex_condition = (DNA_seq_length < 200 or DNA_seq_length > 1000) and (DNA_seq_length % 3 == 0 or DNA_seq_length % 9 == 0)

# Print the results
print(f"DNA sequence length: {DNA_seq_length}")
print(f"Is in range 10 to 100: {in_range_10_to_100}")
print(f"Is lower than 50 or higher than 1000: {lower_than_50_or_higher_than_1000}")
print(f"Meets complex condition: {complex_condition}")

# Optional: Test with different values
test_values = [5, 50, 150, 999, 1001, 1800]

print("\nTesting with different values:")
for value in test_values:
    DNA_seq_length = value
    in_range_10_to_100 = 10 <= DNA_seq_length <= 100
    lower_than_50_or_higher_than_1000 = DNA_seq_length < 50 or DNA_seq_length > 1000
    complex_condition = (DNA_seq_length < 200 or DNA_seq_length > 1000) and (DNA_seq_length % 3 == 0 or DNA_seq_length % 9 == 0)
    
    print(f"\nDNA sequence length: {DNA_seq_length}")
    print(f"Is in range 10 to 100: {in_range_10_to_100}")
    print(f"Is lower than 50 or higher than 1000: {lower_than_50_or_higher_than_1000}")
    print(f"Meets complex condition: {complex_condition}")
````
